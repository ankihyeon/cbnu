```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 로컬에 저장된 모델 경로
local_model_path = "./EXAONE-3.5-7.8B-Instruct"

# 로컬 모델 불러오기
tokenizer = AutoTokenizer.from_pretrained(local_model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(local_model_path, trust_remote_code=True)

print("✅ 로컬 모델 로드 완료!")
```

```python
from llama_cpp import Llama

# 모델 경로와 세부 설정
llm = Llama(
    model_path="./EXAONE-Deep-7.8B-BF16.gguf",
    n_ctx=4096,
    n_gpu_layers=-1,  # 모든 레이어를 GPU에서 실행
    verbose=True      # 실행 로그 확인용 (선택)
)

# 모델 호출
response = llm("너는 무슨 모델이야?")

# 결과 출력
print(response['choices'][0]['text'].strip())
```
